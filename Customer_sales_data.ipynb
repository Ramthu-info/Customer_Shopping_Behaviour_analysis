```python
#
file["Review Rating"] = file.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
file.isnull().sum()
```

```python
file.columns = file.columns.str.lower()
```

```python
file.columns = file.columns.str.replace(" ","_")
```

```python
file = file.rename(columns = { 'customer_id' : 'cust_id', 
                              'purchase_amount_(usd)' : 'purchase_amount',
                              'frequency_of_purchases' : 'freq_of_purchase',
                              'subscription_status' : 'subscribe_status'}) # for simplefy the name for typing then typing code easily
file.columns
```

```python
file['age_group'] = pd.qcut(file['age'],q = 4,labels = [ 'young' ,'Adult','middle_age','senior']) # to group wise divide the age for using this comment
```

```python
file.info()
```

```python
file[['age_group','age']].head(15)
```

```python
# create new column purchase_frequency_days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
file['purchase_freq_days'] = file['freq_of_purchase'].map(frequency_mapping)
```
